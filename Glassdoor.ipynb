```python
## get job title
def get_title():
    job_list=driver.find_elements(By.CSS_SELECTOR,"a[class^='jobLink css-1rd3saf eigr9kq3']")
    job_titles=[job.text for job in job_list]
    return job_titles
```

```python
## get company
def get_company():
    company_list=driver.find_elements(By.CSS_SELECTOR,"a[class^=' css-l2wjgv e1n63ojh0 jobLink']")
    job_company=[comp.text for comp in company_list]
    #len(job_company)
    return job_company
```

```python
## get location
def get_location():
    location_list=driver.find_elements(By.CSS_SELECTOR,"span[data-test^=emp-location]")
    job_location=[loc.text for loc in location_list]
    #len(job_location)
    return job_location
```

```python
##get the links for the ads
def get_link():
    link_list=driver.find_elements(By.CSS_SELECTOR,"a[class^='jobLink css-1rd3saf eigr9kq3']")
    job_link=[link.get_attribute('href') for link in link_list]
    #len(job_link)
    return job_link
```

```python
#get all the the links
#first page
job_link=get_link()
#other pages
for page in pages_list:
    driver.get(f"https://www.glassdoor.com/Job/amsterdam-junior-data-jobs-SRCH_IL.0,9_IC3064478_KO10,21_IP{page}.htm?includeNoSalaryJobs=true&radius=10&pgc=AB4AAoEAPAAAAAAAAAAAAAAAAfTUucgAfQEBARAIHXAqy3rsUGx%2BY1nPO6%2BtZawu0OzIZHJYeQHFSF7xxpuNbja%2F2uoblU1hFf0fU0MgV8TcOI9KmYj6CvRmJEveyoWvUY%2BUVgR8fL%2FPVAOa1W%2B%2Bd383%2Bd%2B0GxLdYhXVFMWpkhWCEu%2Bwr3HVi%2FKWbPY5DKd6XZAyT33mAAA%3D")
    link_list=get_link()
    for link in link_list:
        job_link.append(link)
```

```python
#get all the info for each page of results
#first page
job_titles=get_title()
job_company=get_company()
job_location= get_location()

#other pages
for page in pages_list:
    driver.get(f"https://www.glassdoor.com/Job/amsterdam-junior-data-jobs-SRCH_IL.0,9_IC3064478_KO10,21_IP{page}.htm?includeNoSalaryJobs=true&radius=10&pgc=AB4AAoEAPAAAAAAAAAAAAAAAAfTUucgAfQEBARAIHXAqy3rsUGx%2BY1nPO6%2BtZawu0OzIZHJYeQHFSF7xxpuNbja%2F2uoblU1hFf0fU0MgV8TcOI9KmYj6CvRmJEveyoWvUY%2BUVgR8fL%2FPVAOa1W%2B%2Bd383%2Bd%2B0GxLdYhXVFMWpkhWCEu%2Bwr3HVi%2FKWbPY5DKd6XZAyT33mAAA%3D")
    titles=get_title()
    for title in titles:
        job_titles.append(title)
    companies=get_company()   
    for comp in companies:
        job_company.append(comp)
    locations=get_location()
    for loc in locations:
        job_location.append(loc)

    
    
```

```python
## go to each link and extract second paragraph
text=[]
for i in range(len(job_link)):
#for i in range(3):
    try:
        print(i)
        driver.get(job_link[i])
        i=i+1
        time.sleep(random.randint(1,3))
        parag=driver.find_element(By.TAG_NAME,'p')
        #parag=driver.find_element(By.XPATH,'''//*[@id="JobDesc1007655262442"]/div/div/p''')
        text.append(parag.text)
        # # Click See more.
        # driver.find_element(By.CLASS_NAME,"artdeco-card__actions").click()
        # time.sleep(2)
        # except:
        #     pass
    except:
          text.append("n/a")
          pass
len(text)
text
```

```python
##get the language of each ad

lan_list=[detect(par) for par in text]
```

```python
# Creating the dataframe 
df = pd.DataFrame(list(zip(job_titles,job_company,
                    job_location, text,lan_list,job_link)),
                    columns =['job_title', 'company_name',
                           'company_location','job_desc','lan','link'])
```
