```python
# getting xlsx files from the folder MyProject

#date="2023-01-
path = "Excel_files\Test"
#path="Excel_files\FILTERED_talent_juniordata_2023-01-05.xlsx"
filenames = glob.glob(path + "\*.xlsx")

# for file in filenames:
#     print(pd.read_excel(file))
df_list = (pd.read_excel(file) for file in filenames)

#pd.read_excel(path)

```

```python
# Concatenate all DataFrames
df = pd.concat(df_list, ignore_index=True)
df
```

```python
##FILTER/CLEAN THE DATA
## delete duplicates
filtered_df=df.drop_duplicates(subset=['job_title','company_name'])
```

```python
#exclude ads not in english

filtered_df=df[df['lan']=='en']
```

```python
# EXPORT DATAFRAME TO EXCEL
date=date.today()
filename=f"Excel_files/FILTERED_ALL_ADS_{date}.xlsx"
filtered_df.to_excel(filename,index=False)
```
